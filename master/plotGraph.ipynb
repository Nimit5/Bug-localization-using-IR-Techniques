```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
def plotGraph(name):
    small = df[df['Dataset']==name]
    x = small['Model'].to_list()
    x = [word.split("_")[0] for word in x]
    x1 = np.array(small['MRR'].to_list())
    y1 = np.array(small['MAP'].to_list())


    plt.rcdefaults()
    fig, ax = plt.subplots(figsize =(8, 5))

    N = len(x)
    ind = np.arange(N)
    width = 0.99
    # autolabel(barlist)
    # # axes and labels                                                                                                                                                                                                  
    xtick_marks = ["a", "b", "c","d","e"]
    xtick_names = ax.set_xticklabels(xtick_marks)
    ax.set_xticks(ind)
    xbuffer = 3
    ax.set_xlim(-xbuffer,len(ind)-1+xbuffer)

    ## the bars                                                                                                                                                                                                        
    colors1 = ['orangered','orangered','orangered','orangered','orangered']
    colors2 = ['red','red','red','red','red']
    plt.subplot(1,2,1)
    barlist = plt.bar(x, x1, width, color=colors1)
    # autolabel(barlist)
    plt.title(name.upper())
    plt.xlabel("Models")
    plt.ylabel("MRR")
    plt.show()


    fig, ax = plt.subplots(figsize =(8, 5))

    plt.subplot(1,2,2)
    barlist = plt.bar(x, y1, width, color=colors2)
    # autolabel(barlist)
    plt.title(name.upper())
    plt.xlabel("Models")
    plt.ylabel("MAP")
    plt.show()
```

```python
plotGraph("codec")
```
